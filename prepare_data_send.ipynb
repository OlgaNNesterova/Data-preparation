```python
import pandas as pd
from pathlib import Path
```

```python
data_dir = Path(r'C:\Users\******\для объединения')  # путь к папке, где лежат файлы excel, которые нужно объединить

df = (pd.concat([pd.read_excel(f, skiprows=7).assign(file_name=f.stem)  # сборка всех файлов в один датафрэйм в цикле
                 for f in data_dir.glob("*.xlsx")], ignore_index=True))
```

На следующем шаге хочу к этой таблице добавить информацию о торгпредах из файла "Лист в ДЗ для АД 3". Загружаю данные из файла в отдельный датафрэйм. 

```python
df_torgpred = pd.read_excel(r'C:\Users\******\Лист в ДЗ для АД 3.xlsx')
```

```python
df_torgpred.head()
```

```python
df_torgpred = df_torgpred.dropna()  # удаление строк с пустыми значениями
```

```python
df_torgpred.head()
```

Заметила, что какие-то имена агентов содержат фамилию, имя и отчество, а какие-то - только фамилию и имя. Предполагаю, что это одни и те же люди. Поэтому, чтобы объединить две таблицы по полю с именем агента, в столбцах "Торгпред" и "Агент" нужно оставить только имя и фамилию, новый столбец с именем и фамилией назвать "Торгпред ФИ".

```python
def split_name(full_name):  # функция, которая оставляет имя и фамилию
    x = full_name.split(' ')[0:2]
    return ' '.join(x)
```

```python
df['Торгпред ФИ'] = df.Торгпред.apply(split_name)  # применяем функцию на объединенном датафрэйме
```

```python
df.head()
```

```python
df_torgpred['Торгпред ФИ'] = df_torgpred.Агент.apply(split_name) # применяем функцию на втором датафрэйме
```

```python
df_torgpred.head()
```

Объединяем два датафрэйма по столбцу "Торгпред ФИ"

```python
df_merged = df.merge(df_torgpred, how = 'left', on = 'Торгпред ФИ')
```

```python
df_merged
```

Итоговый датафрэйм сохраняем в виде файла excel

```python
df_merged.to_excel(r'C:\Users\******\final_data.xlsx')
```
